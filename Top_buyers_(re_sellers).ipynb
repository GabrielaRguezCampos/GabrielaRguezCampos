## All emails

```python
# Group by CustomerEmail and calculate total items purchased
orders_by_customer = order_df.groupby('CustomerEmail')['OrderTotalQuantity'].sum().reset_index()
orders_by_customer.columns = ['CustomerEmail', 'TotalItemsPurchased']
#print(orders_by_customer.head())


# Merge with full customer dataset to include company info
reseller_check_df = pd.merge(
    orders_by_customer,
    customer_df[['Email']],
    left_on='CustomerEmail',
    right_on='Email',
    how='left'
)

# Filter for high purchase volume (e.g., more than 100 items)
potential_resellers = reseller_check_df[reseller_check_df['TotalItemsPurchased'] > 100]

# Now sort the filtered result
potential_resellers = potential_resellers.sort_values(by='TotalItemsPurchased', ascending=False)

potential_resellers = potential_resellers.drop(columns=['Email'])

# Show result
potential_resellers.head(10)
```

```python
import matplotlib.pyplot as plt

# Total number of unique customers
total_customers = order_df['CustomerEmail'].nunique()

# Number of potential resellers (already filtered)
num_resellers = potential_resellers['CustomerEmail'].nunique()

# Number of other customers
num_other_customers = total_customers - num_resellers

# Pie chart data
labels = ['Potential Resellers', 'Other Customers']
sizes = [num_resellers, num_other_customers]
colors = ['skyblue', 'lightgray']

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title(f'Distribution of Customer Types (Total: {total_customers})')
plt.axis('equal')  # Equal aspect ratio to ensure circle
plt.show()

# Print the amounts
print(f"Total customers: {total_customers}")
print(f"Potential resellers: {num_resellers}")
print(f"Other customers: {num_other_customers}")
```

## Heat emails

```python
# Define one or more reseller-like email domains
reseller_domains = ['@heat.com']  # Add more if needed

# Filter orders that match any of those domains
reseller_orders = order_df[order_df['CustomerEmail'].str.contains('|'.join(reseller_domains), case=False, na=False)]

# Group by email to see how many items each email purchased
reseller_summary = reseller_orders.groupby('CustomerEmail')['OrderTotalQuantity'].sum().reset_index()
reseller_summary.columns = ['CustomerEmail', 'TotalItemsPurchased']

# Sort results for clarity
reseller_summary = reseller_summary.sort_values(by='TotalItemsPurchased', ascending=False)

reseller_summary_filtered = reseller_summary[reseller_summary['TotalItemsPurchased'] > 200]


pd.set_option('display.max_rows', None)

# Show the result
reseller_summary_filtered.head(10)
```

## Re-sellers by country

```python
# Get list of reseller emails
reseller_emails = potential_resellers['CustomerEmail'].unique()

# Filter customer_df for those emails
reseller_locations = customer_df[customer_df['Email'].isin(reseller_emails)]

# Display relevant columns (e.g., Country, Province, City)
reseller_locations = reseller_locations[['Email', 'CountryName', 'ProvinceName', 'City']]
reseller_locations.head()
```

```python
# Filter order_df for reseller emails
reseller_orders = order_df[order_df['CustomerEmail'].isin(reseller_emails)]

# Select and view shipping location columns
reseller_shipping = reseller_orders[['CustomerEmail', 'ShippingCity', 'ShippingProvinceName', 'ShippingCountryName']]
reseller_shipping.drop_duplicates().head()
```

```python
# Step 1: Filter for US resellers
us_resellers = reseller_locations[reseller_locations['CountryName'] == 'United States']

# Step 2: Count by city and get top 5
top_us_cities = us_resellers['City'].value_counts().head(5)

# Step 3: Print the results
print("Top 5 U.S. cities with most potential resellers:")
for city, count in top_us_cities.items():
    print(f"{city}: {count}")

# Step 4: Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
top_us_cities.plot(kind='bar', color='mediumseagreen')
plt.title('Top 5 U.S. Cities with Most Potential Resellers')
plt.ylabel('Number of Resellers')
plt.xlabel('City')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Count resellers by country and get top 5
reseller_country_counts = reseller_locations['CountryName'].value_counts().head(5)

# Print country counts
print("Top 5 countries with most potential resellers:")
for country, count in reseller_country_counts.items():
    print(f"{country}: {count}")

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
reseller_country_counts.plot(kind='bar', color='skyblue')
plt.title('Top 5 Countries with Most Potential Resellers')
plt.ylabel('Number of Resellers')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Filter out US from the data
reseller_country_counts_excl_us = reseller_locations[reseller_locations['CountryName'] != 'United States']
top_countries_excl_us = reseller_country_counts_excl_us['CountryName'].value_counts().head(5)

# Print counts
print("Top 5 countries (excluding United States):")
for country, count in top_countries_excl_us.items():
    print(f"{country}: {count}")

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
top_countries_excl_us.plot(kind='bar', color='salmon')
plt.title('Top 5 Countries with Most Potential Resellers (Excl. US)')
plt.ylabel('Number of Resellers')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
